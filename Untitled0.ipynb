```python
# Ejercicio N14 Del conjunto de cadenas ["E", "F", "M", "A"] que corresponde a las iniciales de los 4 primeros meses del año, mostrar el nombre completo de dichos meses.

mi_lista=[ "E" ,"F" ,"M" ,"A"]
for i in mi_lista:
  if i== "E" :
    print("Enero")
  elif i== "F":
    print("Febrero")
    i == "M"
    print("Marzo")
    i == "A"
    print("Abril")
```

```python
# Ejercicio N15 Leer un texto y determinar cuántas veces dicho texto contiene cada una de las vocales en minúscula y mayúscula.

texto = input("Ingrese el texto: ")
vocales = {'a': 0, 'e': 0, 'i': 0, 'o': 0, 'u': 0, 'A': 0, 'E': 0, 'I': 0, 'O': 0, 'U': 0}
for letra in texto:
    if letra in vocales:
        vocales[letra] += 1
for vocal, count in vocales.items():
    print(f"{vocal}: {count}")
```

```python
# Ejercicio N17 Leer 20 números enteros y averiguar cuántos son pares y cuántos son impares.

numero_pares = 0
numeros_impares = 0

for i in range(20):
  num = int(input("Ingrese un numero entero: "))
  if num % 2 == 0:
    numeros_pares += 1
else:
  numeros_impares += 1

  print(f"Numeros pares: {numeros_pares}")
  print(f"Numerosimpares: {numeros_impares}")
```
